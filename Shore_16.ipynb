```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student=pd.read_csv('/content/drive/MyDrive/Datasets/sns_data (1).csv')
student
```

# **LinePlot using SeaBorn**

```python
#axes level
sns.lineplot(data=student,x='week',y='attendance_rate',errorbar=None,hue='gender')
```

```python
#figure plot
sns.relplot(kind='line',data=student,x='week',y='attendance_rate',errorbar=None,hue='gender')
```

```python

sns.relplot(kind='line',data=student,x='week',y='test_score',errorbar=None,hue='class_level')
```

# **Scatterplot using seaborn**

```python
sns.scatterplot(data=student,x='study_hours',y='test_score',hue='gender',style='subject',size='Tshirt_size')
```

```python
sns.relplot(kind='scatter',data=student,x='study_hours',y='test_score',hue='gender',style='subject',size='Tshirt_size')
```
