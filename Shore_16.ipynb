# **Facet Plot**

```python
student=pd.read_csv('/content/drive/MyDrive/Datasets/student_dataset_complete.csv')
student.head(5)
```

```python
sns.scatterplot(data=student,x="time_study",y='test_score')
```

```python
sns.relplot(kind='scatter',data=student,x="time_study",y='test_score',col='gender',row='hostel',hue='gender',color=['red','blue'])
```


# **histogram using Seaborn**

# **barplot**

```python
sns.barplot(data=student,x='gender',y='Marks',errorbar=None)
```

```python
sns.barplot(data=student,x='gender',y='Marks',errorbar=None,estimator=np.median)
```
