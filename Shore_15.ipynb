2D line plot from pandas data

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/enrollment_data.csv')
df
```

```python
plt.xlabel('Year')
plt.ylabel('Enrollment')
plt.title('Trend of a students enrollment over years')
plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing')
plt.legend(loc='best')
```

```python
plt.xlabel('Year')
plt.ylabel('Enrollment')
plt.title('Trend of a students enrollment over years')
plt.plot(df['Year'],df['Programming'],label='Programming',color='red',linewidth=3,linestyle='dashed',marker='o',markersize=10)
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing',color='black',linewidth=3,linestyle='dashdot')
plt.legend(loc='best')
plt.grid()
```

# **Scatterplot**





```python
students=pd.read_csv('/content/drive/MyDrive/Datasets/student_IQdata.csv')
```

```python
students
```

```python
plt.xlabel('Study Hours')
plt.ylabel('IQ score ')
plt.title("Iq scores according to study hours")
plt.scatter(students['Study_Hour'],students['IQ_Score'])
```

```python
plt.xlabel('Shoe Size')
plt.ylabel('IQ score ')
plt.title("Iq scores according to shoe size")
plt.scatter(students['Shoe_Size'],students['IQ_Score'])
```

```python
plt.xlabel('Chilling Hour')
plt.ylabel('IQ score ')
plt.title("Iq scores according to chilling hours")
plt.scatter(students['Chilling_Hours'],students['IQ_Score'],marker='o',color='green')
#negative co relation
```

Histogram(data distributed over a factor)

```python
plt.hist(students['IQ_Score'],color='red',edgecolor='black')
plt.xlabel('IQ')
plt.ylabel('Frequency of students')
```

```python
#based on study hours
plt.hist(students['Study_Hour'],color='red',edgecolor='black',bins=20)
plt.xlabel('Study hours')
plt.ylabel('Frequency of students')
```

```python
#based on study hours
plt.hist(students['Study_Hour'],color='red',edgecolor='black',bins=[1,2,3,4,5])
plt.xlabel('Study hours')
plt.ylabel('Frequency of students')
```
