# **Barchart**

```python

data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
gender_group=df.groupby('Gender').size()
gender_group.index
```

```python
gender_group.values
```

```python
plt.bar(gender_group.index,gender_group.values,color=['Orange','Pink'],edgecolor='black')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title("Bar chart of Gender among students")
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/student_data (1).csv')
```

```python
df
```

```python
status=df.groupby('CompletionStatus').size()
plt.bar(status.index,status.values,color=['blue','green','pink'])
plt.xlabel('CompletionStatus')
plt.ylabel('Count')
plt.title("Bar chart of CompletionStatus among students")
```

```python
location=df.groupby('Location').size()
plt.bar(location.index,location.values,color=['blue'])
plt.xlabel('Location')
plt.ylabel('Count')
plt.title("Bar chart of Location among students")
```

# **PieChart**

```python
status=df.groupby('CompletionStatus').size()
plt.pie(status,labels=status.index,autopct='%1.1f%%',explode=(0.1,0,0),shadow=True)
plt.show()
```
